```python
import matplotlib.font_manager as fm

# 扫描系统字体
font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
for font_path in font_list:
    try:
        # 尝试获取字体名称
        prop = fm.FontProperties(fname=font_path)
        name = prop.get_name()
        # 如果名称里包含 'CJK', 'Sans', 'Hei', 'Zen' 等关键词，通常支持中文
        if any(key in name for key in ['CJK', 'SC', 'TC', 'Hei', 'Zen', 'Sun']):
            print(f"找到可用中文字体: {name}  --> 路径: {font_path}")
    except:
        continue
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math, json

data = json.load(open("./outputs/shuffle.json"))
x_value = [int(math.pow(1.5, i)) for i in range(2, 13)]
records = []; keys = [r'$\Delta \mathrm{L}_{shuffle}$', r'$\Delta \mathrm{L}_{original}$', r'$\Delta \mathrm{L}_{locality}$']

df = pd.DataFrame(data, columns=keys)

df["x"] = x_value

# 画图风格（和之前一致）
sns.set_theme(style="white", context="paper", font_scale=1.1)

fig, axes = plt.subplots(1, 3, figsize=(10, 4), sharex=True, dpi=150)

for i, label in enumerate(keys):
    sns.lineplot(
        data=df,
        x="x",
        y=label,
        marker="o",
        markersize=5,
        linewidth=2.5,
        ax=axes[i]
    )
    axes[i].set_xscale("log")
    axes[i].set_title(label, fontsize = 20)
    axes[i].set_xlabel(r"$N_{perm}$", fontsize = 20)
    axes[i].set_ylabel('')
    axes[i].tick_params(axis="y", labelsize=16)
    axes[i].tick_params(axis="x", labelsize=16)

plt.tight_layout()
plt.show()
```

```python
import json
import torch
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

src = json.load(open("./outputs/shuffle_map.json"))

# 假设你有三个 10x10 的方阵 (2D Tensors)
# 这里生成随机数据作为演示
t = 5
matrix1 = torch.tensor(src["ref"])
matrix2 = torch.tensor(src["new"]).div(t).softmax(-1)
matrix3 = torch.tensor(src["old"]).div(t).softmax(-1)

# 可视化列表
matrices = [matrix1, matrix2, matrix3]
titles = ['Permutation Map', 'Shuffled Map', 'Original Map']


my_cmap = LinearSegmentedColormap.from_list("custom", ["#66ccff", "#2950FF"])


# 3. 创建画布，不使用 tight_layout 以便手动控制右侧空间
fig, axes = plt.subplots(1, 3, figsize=(15, 5), dpi = 150)

for i, ax in enumerate(axes):
    # 使用 vmin/vmax 统一刻度，使用 'Blues_r' 避免对比色
    im = ax.imshow(matrices[i].numpy(), cmap=my_cmap, vmin=0, vmax=1)
    ax.set_title(titles[i], fontsize = 20)
    ax.axis('off') # 可选：关闭坐标轴显得更清爽

# 4. 添加公共颜色条
# fraction: 颜色条占原图比例；pad: 与图的间距
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), fraction=0.02, pad=0.04)
cbar.set_label('Ground Truth Probability', fontsize = 20)

plt.show()
```
